```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/processed/advanced_cleaned_data.csv")
df.head()
```

```python
yearly_malaria = df.groupby("year")["malaria_cases"].mean()

plt.figure()
yearly_malaria.plot()
plt.xlabel("Year")
plt.ylabel("Average Malaria Cases")
plt.title("Yearly Trend of Malaria Cases")
plt.show()
```

```python
monthly_malaria = df.groupby("month")["malaria_cases"].mean()

plt.figure()
monthly_malaria.plot(kind="bar")
plt.xlabel("Month")
plt.ylabel("Average Malaria Cases")
plt.title("Monthly Seasonality of Malaria")
plt.show()
```

```python
seasonal_malaria = df.groupby("season")["malaria_cases"].mean()

plt.figure()
seasonal_malaria.plot(kind="bar")
plt.xlabel("Season")
plt.ylabel("Average Malaria Cases")
plt.title("Malaria Cases by Season")
plt.show()
```

```python
plt.figure()
sns.scatterplot(x="avg_temp_c", y="malaria_cases", data=df)
plt.title("Temperature vs Malaria Cases")
plt.show()
```

```python
plt.figure()
sns.scatterplot(x="precipitation_mm", y="malaria_cases", data=df)
plt.title("Rainfall vs Malaria Cases")
plt.show()
```

```python
plt.figure()
sns.scatterplot(x="healthcare_budget", y="malaria_cases", data=df)
plt.title("Healthcare Budget vs Malaria Cases")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(
    df[[
        "avg_temp_c",
        "precipitation_mm",
        "air_quality_index",
        "uv_index",
        "population_density",
        "healthcare_budget",
        "malaria_cases"
    ]].corr(),
    annot=True,
    cmap="coolwarm"
)
plt.title("Correlation Heatmap")
plt.show()
```
